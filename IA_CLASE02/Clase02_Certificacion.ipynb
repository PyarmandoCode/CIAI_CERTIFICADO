## Regresion Lineal
- Es Un metodo estadistico que se utiliza para modelar entre una variable dependiente (o una respuesta) y una o mas variables independientes(o predictores) mediante una ecuacion lineal.
- y = mx + b
  - Donde y : Variable dependiente
  - Donde x : Variable independiente
  - Donde m : Es la pediente de la linea 
  - Donde b : Es la insercion en el eje.
- Ejemplo : Queremos predecir la puntuacion de un estudiante  (Y) en un examen basandose en el numero de horas que estudio (X)
- Es construir una ecuacion matematica que defina los valores de X e Y y poder hacer predicciones con ambos datos
 !["Regresion Lineal"](rlineal.jpg)

```python
#Crear una MODELO de RL que me permita predeccir cuanto obtendra un estudiante 
#de acuerdo a las horas de estudo teniendo como datos de ejemplo obtenidos 
#anteriormente
horas_estudio=[1,2,3,4,5] # Valor de X
puntuacion_de_examen=[60,70,75,85,95] #Valor de Y
#Calcular la regresion lineal manualmente
promedio_horas=sum(horas_estudio)/len(horas_estudio)

promedio_puntuacion=sum(puntuacion_de_examen)/len(puntuacion_de_examen)
m=sum((xi - promedio_horas) * (yi- promedio_puntuacion)for xi,yi, in zip(horas_estudio,
puntuacion_de_examen))/sum((xi-promedio_horas)**2 for xi in horas_estudio)
b=promedio_puntuacion - m * promedio_horas

#predicciones
horas_nuevas=10
prediccion=m*horas_nuevas + b
prediccion

```

## Algoritmos de Inteligencia Artificial
- **Regresion Lineal** .- Se utiliza para predecir una variable continua en funcion a una o mas variables independientes.
    - Ejemplo : Predecir el precio de una casa en funcion a su tamaño y numero de habitaciones.
    - Bibliotecas de Python: scikit-learn('sklearn')
- **Regresion Logistica** .- Se utiliza para problemas de clasificacion binaria 
    - Ejemplo : Como predecir si un correo electronico es spam o no.
    - Biblioteca de Python: from sklearn.linear_model import LogisticRegression
- **Arbol de Desiciones** .- Se puede utilizar para predecir una variable objetivo tomando decisiones basadas en preguntas.
    - Ejemplo : Si una persona es propensa a comprar en un sitio web de acuerdo a su edad ingresos y caracteristicas
    - Biblioteca de Python: from sklearn.tree import DecisionTreeClassifier 
- **Random Forest**.- Un Conjunto de arboles de decision que combina multiples modelos para mejorar la precision y reducir el sobreajuste.
    - Ejemplo : Deteccion de Fraudes de Tarjetas
    - Biblioteca de Python: from sklearn.ensemble import RandomForestClassifier
- **K Vecinos mas cercanos(KNN)** .-Un Algortimo de clasificacion que asigna una etiqueta a un punto de datos basandose en las etiquetas de puntos de datos vecionos.     
    - Ejemplo: Clasificar un nuevo punto de datos basandose en las clase
    - Biblioteca de Python: from sklearn.neighbors import KNeighborsClassifier
- **Redes Neuronales Convolucionales(CNN)**.-Especialmente diseñadas para el procesamiento de imagenes.
    - Ejemplo : Reconocimeinto Facial,clasificacion de imagenes,diagnostico medico a partir de imagenes de resonancia magnetica
     - Bibliotecas de Python: TensorFlow,Keras,PyTorch
     - Keras from keras.models import Sequential
- **Redes Neuronales Recurrentes (RNN)** .- Diseñadas para trabajar con datos secuenciales como series temporales o texto.
    - Ejemplo : Traduccion automatica,generacion de texto,analisis de sentimientos en redes sociales
    - Bibliotecas de Python: TensorFlow,Keras,PyTorch
    - import torch.nn as nn
